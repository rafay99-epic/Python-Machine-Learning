### Reading CSV File of Pre-Processing of data

```python
df2 = pd.read_csv('Pre-Processing-Data.csv', header='infer')
```

```python
# Display the data frame
df2
```

### Word Tokenization

```python
word_tokenizer = nltk.WordPunctTokenizer()
word_tokens = [word_tokenizer.tokenize(sent) for sent in df2.tweets]
word_tokens
```

```python
len(word_tokens)
```

### Applying the Word2Vec Model

```python
model = Word2Vec(word_tokens, min_count = 1, window = 5, sg = 1)
```

### Finding the values for word hate

```python
v1 = model.wv['hate']
print(v1)
```

### Finding the Similar Word

```python
sim_words = model.wv.most_similar('fuck')
print(sim_words)
```

### Saving Model

```python
model.save("word2vec.model")
```
