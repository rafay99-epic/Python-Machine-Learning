```python
df2 = pd.read_csv('Pre-Processing-Data.csv', header='infer')
```

```python
df2
```

```python
corp =df2.tweets 
print(corp)
```

```python
#tokenizing the data
tokenized_sents = [word_tokenize(i) for i in corp]
for i in tokenized_sents:
     print(i)
tokenized_sents
```

```python
#Creating a model
model2 = Word2Vec(tokenized_sents, min_count = 1, window = 5, sg = 1)
```

```python
# word2vec = Word2Vec(tokenized_sents, min_count=2)
# vocabulary = word2vec.wv.vocab
v1 = model2.wv['hate']
# print(v1)
sim_words = model2.wv.most_similar('fuck')
print(sim_words)
```

```python
model2.corpus_count
```

```python
model2.epochs
```

```python
#Training the model
model2.train(tokenized_sents, total_examples=model2.corpus_count,epochs=model2.epochs)
```

```python
#Saving the model
model2.save("word2vec.model")
```

```python
#Finding the similary word and testing the model
model2.wv.most_similar("hate")
```

```python
#finding similarity between two words
model2.wv.similarity(w1="anger", w2="annoyed")
```
