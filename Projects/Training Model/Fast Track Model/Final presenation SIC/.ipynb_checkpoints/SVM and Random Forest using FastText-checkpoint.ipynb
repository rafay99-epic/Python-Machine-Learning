# Class Imbalance Problem

```python
from imblearn.over_sampling import SMOTE       # importing over-sampling (SMOTE) library for class imbalance problem
smote = SMOTE()                                # Defining SMOTE 
```

```python
xtrain_smote, ytrain_smote = smote.fit_resample(a.astype('float'),Data.ClassLabel)        # Dividing embeded dataset and their corresponding class labels
```

```python
from collections import Counter     # Importing Counter for differentiate imbalance and balanced datasets
```

```python
print("Before SMOTE :" , Counter(Data.ClassLabel))     # Imbalanced(Previous dataset)
print("After SMOTE :" , Counter(ytrain_smote))         # Balanced (New Dataset)
```

```python
xtrain_smote              # Displaying embedding of balanced tweets
```

### Visualization of Balanced Dataset

```python
g = sns.countplot(ytrain_smote)
g.set_xticklabels(['Non-Offensive','Offensive'])
plt.show()
```

# Train Test Split

```python
# Dividing data into training and testing data along with their seperate corresponding Class Labels
Train_X, Test_X, Train_Y, Test_Y = train_test_split(xtrain_smote, ytrain_smote,test_size=0.3, random_state=3 )
```

```python
len(Train_Y)        # Displaying Training data
```

# 1.  Support Vectored Machine(SVM)

```python
from sklearn import svm     # Importing SVM library
```

### Training Data

```python
SVM = svm.SVC(C=1.0, kernel='linear', degree=3)   # Defining SVM
SVM.fit(Train_X, Train_Y)                         # Training the dataset(Train_X)
```

```python
# Importing Libraries for Accuracy, Recall, F1 Score and Evaluation Matrix
from sklearn.metrics import f1_score, recall_score, accuracy_score, classification_report, confusion_matrix
```

### Predictions

```python
predictions_SVM=SVM.predict(Test_X)     # Testing data (Test_X) 
```

#### Accuracy

```python
# Using accuracy_score function to get the accuracy
print("SVM Accuracy Score -> ",accuracy_score(predictions_SVM, Test_Y)*100)     
```

```python
# Using recall_score function to get the recall
print(recall_score( predictions_SVM,Test_Y, average="weighted"))
```

```python
# Using f1_score function to get the F1 Score
print(f1_score(Test_Y, predictions_SVM, average = 'weighted'))
```

# Evaluation Matrix

```python
print(classification_report(Test_Y, predictions_SVM))       # Evaluation Matrix
```

# Confusion Matrix

```python
# View confusion matrix for test data and predictions
confusion_matrix(Test_Y, predictions_SVM)
```

```python
# Get and reshape confusion matrix data
matrix = confusion_matrix(Test_Y, predictions_SVM)
matrix = matrix.astype('float') / matrix.sum(axis=1)[:, np.newaxis]

# Build the plot
plt.figure(figsize=(16,7))
sns.set(font_scale=1.4)
sns.heatmap(matrix, annot=True, annot_kws={'size':10},
            cmap=plt.cm.Greens, linewidths=0.2)

# Add labels to the plot
class_names = ['Non-Offensive', 'Offensive']
tick_marks = np.arange(len(class_names))
tick_marks2 = tick_marks + 0.5
plt.xticks(tick_marks, class_names, rotation=25)
plt.yticks(tick_marks2, class_names, rotation=0)
plt.xlabel('Predicted label')
plt.ylabel('True label')
plt.title('Confusion Matrix for Random Forest Model')
plt.show()
```

# 2. Random Forest

```python
from sklearn.model_selection import cross_val_score , RepeatedStratifiedKFold   
from sklearn.ensemble import RandomForestClassifier    # Importing Random Forest Classifier
```

### Training with Random Forest

```python
rf = RandomForestClassifier(n_estimators=400, random_state=15,verbose=1).fit(Train_X, Train_Y)    # training data using Train_X
```

### Prediction using Random Forest

```python
prediction = rf.predict(Test_X)     # Testing using Test_X
```

### Applying 10-Folds Cross Validation

```python
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)     # Defining K-Folds Cross Validation model
```

```python
n_scores = cross_val_score(rf, Train_X, Train_Y, scoring='accuracy', cv=cv, n_jobs=-1, error_score='raise', verbose=2)
# Computing mean Accuracy of 10 Folds Cross Validation
print('Accuracy: %.3f (%.3f)' % (np.mean(n_scores), np.std(n_scores)))
```

#### Accuracy

```python
print(accuracy_score(Test_Y, prediction))     # Computing Accuracy
```

```python
print(recall_score(Test_Y, prediction, average='micro'))   #Computing Recall
```

```python
print(f1_score(Test_Y, prediction, average = 'micro'))    # Computing F1 Score
```

# Evaluation Matrix

```python
print(classification_report(Test_Y, prediction))      # Displaying Evaluation Matrix
```
