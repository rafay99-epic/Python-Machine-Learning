## Box Plot

```python
x = np.random.randn(10000)*10+3
y = np.random.randn(10000)*10+5
z = np.random.randn(10000)*10+1
plt.boxplot([x,y,z],0)
plt.title('Box Plot')
plt.show()
```

```python
plt.boxplot([x,y,z],vert = False)
plt.title('Box Plot')
plt.show()
```

```python
x= np.linspace(0,10,100)
y  = np.sin(x)
plt.plot(x,y,color='orange', linewidth =2)
plt.xlabel('x')
plt.ylabel('sin')
plt.title('Line Plot')
plt.show()
```

```python
x= np.linspace(0,10,100)
y  = np.sin(x)
plt.plot(x,y,color='blue', linewidth =2, linestyle='-.')
plt.xlabel('x')
plt.ylabel('sin')
plt.title('Line Plot')
plt.show()
```

## Scatter             Plot

```python
x = np.linspace(0,10,100)
y = np.tan(x)
plt.plot(x,y,color='red',marker='*',linestyle='none',markersize=5, alpha=0.5)
plt.xlabel('X')
plt.ylabel('Sin')
plt.title('SCATTER PLOT')
plt.show()
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
plt.plot(x,y,color='#FF8C00',marker='o',linestyle='none',markersize=10, alpha=0.5)     # RGB color.
plt.xlabel('X')
plt.ylabel('Sin')
plt.title('SCATTER PLOT')
plt.show()
```

### exercise 210

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.linspace(0,10,100)
```

```python
y= np.sin(x)
z= np.cos(x)
```

```python
fig0 =plt.figure()
axes0=fig0.add_axes([0,0,1,1])
axes0.plot(x,y,color='red',linestyle = '-')
axes0.set_xlabel('X')
axes0.set_ylabel('Y')
axes0.set_title('Line Plot')
plt.show()
```

```python
fig0 = plt.figure()
axes0 = fig0.add_axes([0,0,1,1])
axes0.plot(y,z,color='purple', linestyle ="-")
axes0.set_xlabel('X')
axes0.set_ylabel('Y')
axes0.set_title('Line Plot')
axes0.set_xlim([0,5])
axes0.set_ylim([-2,2])    
plt.show()
```

```python
fig0 = plt.figure(figsize=(10,1))
axes0 = fig0.add_axes([0,0,1,1])
axes0.plot(y,z,color= 'red', linestyle ='-')
axes0.set_xlabel('X')
axes0.set_ylabel('Y')
axes0.set_title('LINE PLOT')
plt.show()
```

```python
fig0=plt.figure(figsize=(8,3), dpi=80)       # Width, height and DPI setting.
axes1 = fig0.add_axes([0,0,0.3,1])            # Left, bottom, width, height.
axes2 = fig0.add_axes([0.5,0,0.3,1])          # Left, bottom, width, height.
axes1.plot(x,y,color='red',linestyle=':')  
axes2.plot(x,z,color='blue',linestyle='-.')
axes1.set_xlabel('X')
axes1.set_ylabel('Y')
axes1.set_title('SINE')
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.set_title('COSINE')
plt.show()
```

```python
fig0=plt.figure(figsize=(8,4), dpi=80)          # Width, height and DPI setting.
axes1 = fig0.add_axes([0,0.5,0.2,0.5])           # Left, bottom, width, height.
axes2 = fig0.add_axes([0.5,0,0.2,0.5])           # Left, bottom, width, height.
axes1.plot(x,z,linestyle='none',marker='^',markersize=10,markerfacecolor='purple',markeredgecolor='red',alpha=0.5)
axes2.plot(x,y,color='blue',linestyle='--')  
axes1.set_xlabel('X')
axes1.set_ylabel('Y')
axes1.set_title('COSINE')
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.set_title('SINE')
plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(6,6))
# (0,0)
c = np.random.randn(10000)
axes[0,0].hist(c,bins=100,color='green',density=True,alpha=0.3)
axes[0,0].set_title('HISTOGRAM')
# (1,0)
axes[1,0].plot(x,z,linestyle='none',marker='^',markersize=10,markerfacecolor='purple',markeredgecolor='red',alpha=0.5)
axes[1,0].set_xlabel('X')
axes[1,0].set_ylabel('Y')
axes[1,0].set_title('COSINE')
# (0,1)
a = np.array(['Q1', 'Q2', 'Q3','Q4'])
b = np.array([ 234.0, 254.7, 144.6, 317.6])
axes[0,1].bar(a,b,color='blue',alpha=0.5)
axes[0,1].set_title('BAR PLOT')
# (1,1)
axes[1,1].plot(x,y,color='green',linestyle=':')  
axes[1,1].set_xlabel('X')
axes[1,1].set_ylabel('Y')
axes[1,1].set_title('SINE')

plt.tight_layout()                             # Avoid overlapping.
plt.show()
```
