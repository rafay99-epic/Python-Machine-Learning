```python
#  A function that returns the strings where the pattern is matched.
# The pattern is input as a regular expression.
def grep(reg_exp, my_text):
    # this is an empty list that will save the result
    result = []
    # this will complie the reg
    reg_exp_compiled = re.compile(reg_exp)
    # this will compare each string and id it is saved or matched then it will be stored in the list called result
    for i in range(len(my_text)):
        if reg_exp_compiled.search(my_text[i]):
            result.append(my_text[i])
    # this will return the values or the output
    return result
```

```python
grep('hate', ['I love you', 'I hate you', 'I like you'])
```

```python
grep("^[0-9]", ["1000 that is!", "you are the number 212", "my age is 21"])
```

```python
# this will complie the reg 
reg_ex = re.compile("([0-9]+)")
# this will be searching for the required answer
m = reg_ex.search("I am 22 years old. She is 20 years old.")
print(m.group(1))
```

```python

reg_ex = re.compile("([0-9]+)[^0-9]+([0-9]+)")
m = reg_ex.search("I am 22 years old. She is 20 years old.")
print(m.group(1))
print(m.group(2))
```

```python
reg_ex = re.compile("([^0-9]+)([0-9]+)[^0-9]+([0-9]+)[^0-9]+([0-9]+)")
m = reg_ex.search("John Wheeler 010-1234-5678")
print((m.group(1)).strip())
print(m.group(2))
print(m.group(3))
print(m.group(4))
```

```python
reg_ex = re.compile("(\d+)")
m = reg_ex.search("I am 22 years old. She is 20 years old.")
print(m.group(1))
```

```python
reg_ex = re.compile("(\d+)\D+(\d+)")
m = reg_ex.search("I am 22 years old. She is 20 years old.")
print(m.group(1))
print(m.group(2))
```

```python

reg_ex = re.compile("(\D+)(\d+)\D+(\d+)\D+(\d+)")
m = reg_ex.search("John Wheeler 010-1234-5678")
print((m.group(1)).strip())
print(m.group(2))
print(m.group(3))
print(m.group(4))
```

```python
# Hide the phone number.
reg_ex = re.compile("(\D+)(\d+)\D+(\d+)\D+(\d+)")
m = reg_ex.search("John Wheeler 010-1234-5678")
print((m.group(1)).strip() + " " + m.group(2) + "-****-****")
```

```python
# Extract only the phone number.
reg_ex = re.compile("(\D+)((\d+)\D+(\d+)\D+(\d+))")
m = reg_ex.search("John Wheeler 010-1234-5678")
print("PHONE : " + m.group(2))
```
