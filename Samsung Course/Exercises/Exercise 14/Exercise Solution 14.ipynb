```python
def getAllParagraphs(filename):
    a_doc = docx.Document(filename)
    fullText = []
    for a_para in a_doc.paragraphs:
        fullText.append(a_para.text)
    return '\n'.join(fullText)
```

```python
print(getAllParagraphs('Artificial Intelligence.docx'))
```
