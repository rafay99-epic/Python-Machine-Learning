```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
import warnings
%matplotlib inline
warnings.filterwarnings('ignore')
```

```python
dat = sns.load_dataset('mpg')
```

```python
dat.shape
```

```python
dat.head(5)
```

```python
sns.distplot(dat.mpg, kde= False, rug= False, bins=50, color='red')
plt.show()
```

```python
sns.displot(dat.mpg, kde = True, bins = 50, color= 'blue')
plt.show()
```

```python
sns.displot(dat.mpg, kde= False, rug= True, bins= 50, color='orange')
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
import warnings
%matplotlib inline
warnings.filterwarnings('ignore')


sns.jointplot(x='weight', y = 'mpg', data= dat, color= 'red', kind= 'scatter')
plt.show()
```

```python

sns.jointplot(x='weight', y = 'mpg', data= dat, color= 'red', kind= 'reg')
plt.show()
```

```python
sns.lmplot(data= dat, x= 'weight', y = 'mpg')
plt.show()
```

```python
sns.lmplot(data= dat, x= 'weight', y= 'mpg', hue='origin')
plt.show()
```

```python
sns.lmplot(data=dat, x='weight', y='mpg', row = 'origin', hue = 'origin', markers=['o','v','^'])
plt.show()
```

```python
sns.lmplot(data= dat, x= 'weight', y= 'mpg', col='origin', markers=['o','^','v'], hue= 'origin',scatter_kws={'s':40},aspect= 0.5, size=5)
plt.show()
```

```python
sns.jointplot(x='weight', y='mpg', data=dat, color='green', kind='kde')
plt.show()
```

```python
sns.barplot(x= 'origin', y= 'mpg', data= dat, estimator= np.sum)
plt.show()
```

```python
sns.barplot(x= 'origin', y = 'mpg', data=dat)
plt.show()
```

```python
sns.countplot(x='origin',data=dat,palette='coolwarm')
plt.show()
```

```python
sns.countplot(x= 'origin',data=dat, hue='origin')
plt.show()
```

```python
sns.boxplot(y='mpg', data=dat, color='green')
plt.show()
```

```python


# Horizontal boxplot.
sns.boxplot(x='mpg', data=dat, color='pink')
plt.show()

```

```python


# Multiple of boxplots.
sns.boxplot(x='origin', y='mpg', data=dat, palette='coolwarm', notch=True)
plt.show()

```
