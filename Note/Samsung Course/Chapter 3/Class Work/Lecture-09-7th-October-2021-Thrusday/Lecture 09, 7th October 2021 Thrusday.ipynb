```python
data = {'Name': ['Ali','Nasir','Moin','Asad','Hassan'],'A': [5,6,9,3,8], 'B': [3.1,9.5,7.2,6.7,7.8], 'C': [25,27,51,63,34],'D':[71,67,78,44,31]}
```

```python
data2 ={'Name': ['Noor', 'Moin','Tahir','Asad','Umar'],'D':[11,9,8,7,5],'X':[3.5,8.6,7.2,8.9,9.3],'Y':[3.5,44.9,53.8,71.2,66.4],'Z':[210,510,339,418,389]}
```

```python
df1= pd.DataFrame(data)
```

```python
df2= pd.DataFrame(data2)
```

```python
df1
```

```python
df2
```

Inner Join

```python
pd.merge(df1, df2, left_on='Name', right_on = 'Name', how='inner')
```

```python
pd.merge(df1, df2, left_on='Name', right_on = 'Name', how='left')
```

```python
pd.merge(df1, df2, left_on='Name', right_on = 'Name', how='right')
```

```python
df3=pd.merge(df1, df2, left_on='Name', right_on = 'Name', how='outer')
df3
```

```python
# this null functin will relace the no value with true and present data will be false
df3.isnull()
```

```python
# this will tell the missue values and what is the count of them who manny values are missing
df3.isnull().sum(axis=0)
```
